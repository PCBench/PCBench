```python
import csv
import os

def write_csv(data, file_name):

    with open(file_name, 'w', newline='') as csvfile:
        spamwriter = csv.writer(csvfile, delimiter=',')
        for row in data:
            spamwriter.writerow(row)

def read_csv(file_name):

    data = []
    with open(file_name) as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        for row in reader:
            data.append(row)
    return data
```

```python
import os
from scipy.spatial.distance import euclidean

from Scripts.Data_extraction.thirdparty.kicad_parser.kicad_pcb import *
from Scripts.Data_extraction.thirdparty.kicad_parser.sexp_parser import *


pcbs = list(os.listdir("../PCBs/"))
wirelength = dict()
via = dict()
for pname in pcbs:
    if pname == "master_metadata.json":
        continue
    pcb = KicadPCB.load(os.path.join("../PCBs/", pname, "processed.kicad_pcb"))
    wl = 0
    for seg in pcb.segment:
        wl += euclidean(seg.start, seg.end)
    wirelength[pname] = wl
    via[pname] = len(pcb.via)

print(via)
print(wirelength)
```

```python
data = read_csv("./Results_FR_PcbRouter_MCTS.csv")
for i, row in enumerate(data):
    if row[13] == "":
        row[13] = "Yes"
        row[14] = wirelength[row[0]]
        row[15] = via[row[0]]
        row[16] = "N/A"
        data[i] = row

write_csv(data, "Results_all.csv")
```
