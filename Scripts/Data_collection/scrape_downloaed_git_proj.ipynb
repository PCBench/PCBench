```python
# extract kicad files from github pcb projects
import os
import json
import shutil
import datetime

# folder store all meta info of downloaded projects
github_meta_folder = '../../github_meta/'
# folder of all downloaded data
all_projs = list(os.listdir('../../benchmarks'))

for git_idx in range(6, 35):
    git_info_json = f"github_{git_idx}.json"
    git_path = os.path.join(github_meta_folder, git_info_json)
    with open(git_path) as json_file:
        proj_infos = json.load(json_file)
    
    for proj_info in proj_infos["items"]:
        proj_name = proj_info['name']
        # file creation
        if proj_name in all_projs:
            c_timestamp = os.path.getctime(os.path.join('../../benchmarks/', proj_name))
            # convert creation timestamp into DateTime object
            c_datestamp = datetime.datetime.fromtimestamp(c_timestamp)
            meta = {
                "org": proj_info["owner"]["html_url"],
                "source": proj_info["html_url"],
                "author": proj_info["owner"]["login"],
                "retrieved at": str(c_datestamp),
                "raw": "",
                "cleaned": "",
                "json": "",
                "supplementary files": "",
                "licenses": proj_info["license"],
                "layers": "",
                "CAD version": ""
            }

        
            print(f"Processing {proj_name}")
            kicad_pcbs = []
            # os.walk() returns subdirectories, file from current directory and 
            # And follow next directory from subdirectory list recursively until last directory
            for root, dirs, files in os.walk(os.path.join('../../benchmarks/', proj_name)):
                for file in files:
                    if file.endswith(".kicad_pcb"):
                        kicad_pcbs.append(os.path.join(root, file))
            # print(kicad_pcbs)
            for kicad_pcb in kicad_pcbs:
                pcb_folder_name = proj_name + "_" + kicad_pcb.split("/")[-1][:-10]
                # print(pcb_folder_name)
                store_folder = os.path.join("../../new_PCBs", pcb_folder_name)
                if not os.path.exists(store_folder):
                    os.mkdir(store_folder)
                shutil.copy(kicad_pcb, os.path.join(store_folder, "raw.kicad_pcb"))
            
                meta["raw"] = os.path.join("../../PCBs", pcb_folder_name, "raw.kicad_pcb")
                meta["cleaned"] = os.path.join("../../PCBs", pcb_folder_name, "processed.kicad_pcb")
                meta["json"] = os.path.join("../../PCBs", pcb_folder_name, "metadata.kicad_pcb")
                try:
                    meta_path = os.path.join(store_folder, "metadata.json")
                    with open(meta_path, "w", encoding='utf-8') as f:
                        metaf = json.dumps(meta, indent=2)
                        f.write(metaf)
                except:
                    print(proj_name)

        else:
            print(f"---------- No {proj_name}---------")
```
